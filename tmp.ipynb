```python
import random
li = [1,2,3,4]

def make_None(input_list, none_count):
    li33= input_list
    indices_to_change = random.sample(range(0,len(input_list)-1), none_count)
    # print(indices_to_change)
    for index in indices_to_change:
        li33[index] = None
    return li33

make_None(li,3)
```

```python
def generate_unique_numbers_in_range(start, end, count):
    if count > (end - start + 1):
        return "Error: Count exceeds the range."
    
    unique_numbers = random.sample(range(start, end + 1), count)
    
    return unique_numbers
generate_unique_numbers_in_range(0,5,5)
```

```python
import pandas as pd

def create_dataframe_with_equal_length(dict_data):
    max_length = max(len(values) for values in dict_data.values())
    filled_data_dict = {key: values + [None] * (max_length - len(values)) for key, values in dict_data.items()}
    df = pd.DataFrame(filled_data_dict)
    return df

# Example usage
your_dict = {
    "column1": [1, 2, 3],
    "column2": [4, 5],
    "column3": [6]
}

resulting_dataframe = create_dataframe_with_equal_length(your_dict)
print(resulting_dataframe)
```
